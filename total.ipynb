# Общие тенденции изучения языков программирования за 2021-2024 года
---

```python
import pandas as pd
import matplotlib.pyplot as plt

# Данные
years = [21, 22, 23, 24]
languages = ['Паскаль', 'Питон', 'C#', 'C++', 'Java']

popularity_data = {
    'Паскаль': [92.31, 98.55, 80.82, 80.58],
    'Питон': [84.62, 95.65, 90.41, 99.28],
    'C#': [38.46, 43.48, 16.44, 64.03],
    'C++': [30.77, 46.38, 28.77, 45.32],
    'Java': [23.88, 31.88, 6.85, 40.29]
}

level_data = {
    'Паскаль': [5.5, 4.72, 4.17, 3.45],
    'Питон': [6.73, 6.82, 6.17, 6.78],
    'C#': [7.4, 4.17, 5.58, 2.94],
    'C++': [5.0, 2.78, 3.29, 1.7],
    'Java': [6.0, 2.14, 4.4, 1.2]
}

count_data = {
    '1': [2, 3, 15, 19],
    '2': [6, 32, 37, 29],
    '3': [1, 6, 10, 33],
    '4': [2, 7, 11, 8],
    '5': [2, 21, 0, 50]
}

# Создаем DataFrame
popularity_df = pd.DataFrame(popularity_data, index=years)
level_df = pd.DataFrame(level_data, index=years)
count_df = pd.DataFrame(count_data, index=years)

# Просмотр данных
popularity_df, level_df, count_df
```

```python
# Построим график изменения популярности по языкам по годам
plt.figure(figsize=(10, 6))
for language in languages:
    plt.plot(popularity_df.index, popularity_df[language], label=language)

plt.title('Изменение популярности языков по годам')
plt.xlabel('Год')
plt.ylabel('Популярность (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Построим график изменения уровня владения по языкам
plt.figure(figsize=(10, 6))
for language in languages:
    plt.plot(level_df.index, level_df[language], label=language)

plt.title('Изменение уровня владения языками по годам')
plt.xlabel('Год')
plt.ylabel('Средний уровень владения')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Строим график изменения общего уровня владения по годам
overall_level = level_df.mean(axis=1)
plt.figure(figsize=(10, 6))
plt.plot(level_df.index, overall_level, label='Общий уровень владения', color='orange', marker='o')

plt.title('Изменение общего уровня владения языками по годам')
plt.xlabel('Год')
plt.ylabel('Средний уровень владения')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Создаем DataFrame для count_data
count_df = pd.DataFrame(count_data, index=years)

# Для каждого года находим общее количество студентов
total_students = count_df.sum(axis=1)

# Преобразуем значения в проценты
percent_df = count_df.div(total_students, axis=0) * 100

# Построим график
plt.figure(figsize=(10, 6))
percent_df.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')

plt.title('Процентное распределение студентов по числу изучаемых языков по годам')
plt.xlabel('Год')
plt.ylabel('Процент студентов (%)')
plt.xticks(rotation=0)
plt.legend(title="Количество языков", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```
